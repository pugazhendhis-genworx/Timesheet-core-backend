```python
from google import genai
from google.genai import types

# ---- CONFIG ----

API_KEY = "YOUR_GMAIL_API_KEY"
FILE_PATH = "/home/pugazhendhi/Documents/TimeIQ/Timesheet-core-backend/attachments/1000259068.jpg"  # noqa  # example: "1000259068.jpg"

# ---- INIT CLIENT ----

client = genai.Client(api_key=API_KEY)

prompt = """
Extract readable text from this timesheet document.
Return only extracted text.
"""

# ---- READ FILE ----

with open(FILE_PATH, "rb") as f:
    file_bytes = f.read()

    # ---- CALL GEMINI ----

    response = client.models.generate_content(
        model="gemini-2.5-flash",  # better for document extraction
        contents=[
            prompt,
            types.Part.from_bytes(data=file_bytes, mime_type="image/jpeg"),
        ],
    )

    print("RAW RESPONSE:")
    print(response)

    print("\nEXTRACTED TEXT:")
    print(response.text)
```

```python
import logging

import pandas as pd

logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)


def parse_excel_timesheet(file_path: str) -> str:
    try:
        df = pd.read_excel(file_path)
        return df.to_string()
    except Exception as e:
        logger.exception("Excel parsing failed")
        raise RuntimeError(f"Excel parse error: {str(e)}") from e


# ask user for file path
file_path = input("Enter the Excel timesheet file path: ")

print("\nParsing Excel...\n")

result = parse_excel_timesheet(file_path)

print("Extracted Data:\n")
print(result)
```
